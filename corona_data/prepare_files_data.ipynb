```python
import pandas as pd
```

```python
train_dataset = pd.read_csv('corona_data/Corona_NLP_train.csv',encoding = "ISO-8859-1")
train_dataset.rename(columns={"OriginalTweet":"text", "Sentiment":"label"}, inplace=True)
train_dataset['label'] = train_dataset['label'].replace(['Neutral', 'Positive', 'Extremely Negative', 'Negative',
       'Extremely Positive'],[0,1,2,3,4]) 
```

```python
point_tts = int(0.8*len(train_dataset))
train_dataset[:point_tts].to_csv('corona_data/train_data.csv')
train_dataset[point_tts:].to_csv('corona_data/eval_data.csv')
```
